```python
# FULL FITNESS COACH LOOP POWERED BY OPENAI API (TEXT AND IMAGE INPUTS. TEXT AND IMAGE SYSTEM PROMPTS)

from openai import OpenAI
from dotenv import load_dotenv, dotenv_values
import os

# Load environment variables from .env file
load_dotenv()  # You can specify a path like load_dotenv("/path/to/your/.env") if needed

# Retrieve the API key from the environment variable
client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))

def chat_loop():
    
    ####Set system prompt, #describe initial image context, and #feed in initial image context
    System_Prompt = "You are Mr. Peak, a personal fitness trainer that provides body building advice based on people's physique photos. There is a seperate tool that handles the image URL's that they provide. Therefore you can accept image URL's Assign your clients a score between 1-100 with 100 being the most muscular. You tell your clients what they need to work on: for example what muscle groups to target and what workouts they need to do to target those muscles. If they do not want to upload a photo of themselves they can just chat. Now introduce yourself!"
    initial_image_prompt = "This text is the system prompt so you are technically not speaking with the user yet: This is what a 100 Physique looks like:" 
    initial_image_URL = 'https://source.boomplaymusic.com/group10/M00/04/06/2cc396baf0294a8889b27a659d647681_464_464.jpg'
    
    Master_messages= [{"role": "user", "content": [{"type": "text", "text": System_Prompt}]}]
    
    Master_messages.append({"role": "user", "content": [{"type": "text", "text": initial_image_prompt}, 
                           {"type": "image_url", "image_url": {"url": initial_image_URL}}]})
    
    # Initialize the conversation with the initial prompt
     
    initial_AI_Generate = client.chat.completions.create(
                model="gpt-4o",
                messages= Master_messages)

    
    System_Prompt_Response = initial_AI_Generate.choices[0].message.content
    Master_messages.append({"role": "user", "content": [{"type": "text", "text": System_Prompt}]})
    Master_messages.append({"role": "assistant", "content": [{"type": "text", "text": System_Prompt_Response}]})
    
    print(f"\nMr. Peak: {System_Prompt_Response}")

    conversation_running = True
    user_input = ""
    
    
    while conversation_running:
        
        
        user_input = input("User: ")

        if user_input.lower() in ["exit", "quit"]:
            conversation_running = False
            print("Ending conversation.")
            break
        
        #Check if there is a link in the initial User Input:
       
        Link_check_dict=([{"role": "user", "content": [{"type": "text", "text": "If there is a URL, filepath, or Link in this input that ends in .jpg or .png respond with 'True.' If there is no URL or file path respond with 'False:'" + user_input}]}])
        
        link_check = client.chat.completions.create(
            model="gpt-4o", messages= Link_check_dict )
        
        Check_Check = link_check.choices[0].message.content
        
     
        #Actually extract the link
        
        
        link_extract_dict =([{"role": "user", "content": [{"type": "text", "text": "If there is a URL, filepath, or Link in this input that ends in .jpg or .png, print only that URL, filepath, or Link back to me. Do not say anything else: " + user_input}]}])
        
        Link_extract = client.chat.completions.create(
            model="gpt-4o", messages= link_extract_dict )
        
        extracted_link = Link_extract.choices[0].message.content
        
        
        
        # Add the user input to the list of messages (Either JUST text prompt or text and images)
        
        if Check_Check == 'True':
            
            Master_messages.append({
                "role": "user",
                "content": [
                    {"type": "text", "text": user_input},
                    {"type": "image_url", "image_url": {"url": extracted_link}}
                ]
            })
            
            AI_Generate = client.chat.completions.create(
                model="gpt-4o",
                messages=Master_messages
            )
        else:
            Master_messages.append({"role": "user", "content": [{"type": "text", "text": user_input}]})
            
            AI_Generate = client.chat.completions.create(
                model="gpt-4o",
                messages=Master_messages
            )
        
        ai_response = AI_Generate.choices[0].message.content
        
        # Print the AI's response
        print(f"Mr. Peak: {ai_response}")
        
        # Add the AI's response to the list of messages
        Master_messages.append({"role": "assistant", "content": [{"type": "text", "text": ai_response}]})
        
        # Print the history of messages (optional)
        #print(f"Messages: {messages}")

# Start the game loop
print("To exit the conversation, type 'exit' or 'quit' ")
chat_loop()
```
